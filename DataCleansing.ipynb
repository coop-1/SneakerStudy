```python
import pandas as pd
```

```python
sneakers_df = pd.read_csv('SneakerData.csv')
income_df = pd.read_csv('US_Income_Data.csv')
```

```python
income_df = income_df[income_df['Description'] == 'Per capita personal income (dollars) 2/']
income_df = income_df.transpose()

```

```python
income_df.columns = income_df.iloc[1]
income_df = income_df.drop(['GeoFIPS'])
income_df = income_df.drop(['GeoName'])
income_df = income_df.drop(['Region'])
income_df = income_df.drop(['TableName'])
income_df = income_df.drop(['LineCode'])
income_df = income_df.drop(['IndustryClassification'])
income_df = income_df.drop(['Description'])
income_df = income_df.drop(['Unit'])
income_df.head()
```

```python
income_df['Year'] = income_df.index.str[:4]
income_df["Year"] = income_df["Year"].astype(int)
income_df = income_df[income_df['Year'] > 2016]
income_df.to_csv('IncomeData_Cleaned.csv', index = True)
```
