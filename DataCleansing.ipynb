```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
sneakers_df = pd.read_csv('SneakerData.csv')
income_df = pd.read_csv('US_Income_Data.csv')
```

```python
income_df = income_df[income_df['Description'] == 'Per capita personal income (dollars) 2/']
income_df = income_df.rename(columns = {'GeoName':'State'})
del income_df['GeoFIPS']
del income_df['Region']
del income_df['TableName']
del income_df['LineCode']
del income_df['IndustryClassification']
del income_df['Unit']
del income_df['Description']
```

```python
income_df = income_df.replace({'State': 'Alaska *'}, "Alaska")
income_df = income_df.replace({'State': 'Hawaii *'}, "Hawaii")
income_df
```

```python
#####################FINAL INCOME DF#####################

income_df = income_df[['State','2017:Q1','2017:Q2','2017:Q3','2017:Q4','2018:Q1','2018:Q2','2018:Q3','2018:Q4','2019:Q1','2019:Q2','2019:Q3']]

income_df = income_df.reset_index()

del income_df['index']
income_df = income_df.melt(id_vars=['State'], value_vars=['2017:Q1','2017:Q2','2017:Q3','2017:Q4','2018:Q1','2018:Q2','2018:Q3','2018:Q4','2019:Q1','2019:Q2','2019:Q3'])

income_df = income_df.rename(columns = {'variable':'Quarter','value':'Income'})

income_df


# new_df
```

```python
income_df_total = income_df
income_df_total = income_df_total.groupby('State').agg({'Income':'mean'})

income_df_total.Income = income_df_total.Income.round()
income_df_total
income_df_total.to_dict('index')
```

```python
sneakers_df['Order Date'] = pd.to_datetime(sneakers_df['Order Date'])
sneakers_df
```

```python
sneakers_df['Month'] = pd.DatetimeIndex(sneakers_df['Order Date']).month
sneakers_df['Year'] = pd.DatetimeIndex(sneakers_df['Order Date']).year
sneakers_df['Sale Price']=(sneakers_df['Sale Price'].replace( '[\$,)]','', regex=True )).astype(int)
sneakers_df['Retail Price']=(sneakers_df['Retail Price'].replace( '[\$,)]','', regex=True )).astype(int)



# sneakers_df = sneakers_df[sneakers_df['Year'] == 2018]


sneakers_df_totaldata = sneakers_df

sneakers_df_totaldata = sneakers_df_totaldata[sneakers_df_totaldata['Year'] == 2018]

sneakers_df_quarter = sneakers_df

sneakers_df_totaldata
```

```python
bins = [0,3,6,9,12]
group_labels = ['Q1', 'Q2', 'Q3', 'Q4']
sneakers_df_quarter['Quarter'] = pd.cut(sneakers_df_quarter["Month"], bins, labels=group_labels)
sneakers_df_quarter = sneakers_df_quarter.rename(columns = {'Buyer Region':'State'})


sneakers_df_quarter['Year'] = sneakers_df_quarter['Year'].astype(str)
sneakers_df_quarter["Year_Quarter"] = sneakers_df_quarter[['Year', 'Quarter']].apply(lambda x: '_'.join(x), axis=1)

sneakers_df_quarter = sneakers_df_quarter.groupby(['Year_Quarter','State']).agg({'Sale Price':'sum'})

sneakers_df_quarter
```

```python
sneakers_df_quarter = sneakers_df_quarter.reset_index()
sneakers_df_quarter
```

```python
bins = [0,3,6,9,12]
group_labels = ['Q1', 'Q2', 'Q3', 'Q4']
sneakers_df['Quarter'] = pd.cut(sneakers_df["Month"], bins, labels=group_labels)
```

```python
sneakers_df['Year'] = sneakers_df['Year'].astype(str)
sneakers_df["Year:Quarter"] = sneakers_df[['Year', 'Quarter']].apply(lambda x: ':'.join(x), axis=1)

sneakers_df = sneakers_df.groupby(['Year:Quarter','Brand','Buyer Region']).agg({
    'Sale Price': ['mean','sum'], 
    'Retail Price': 'mean',
})
```
